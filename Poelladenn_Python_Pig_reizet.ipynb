## C'hoari ar "PIG"

C'hoari ar "Pig" a zo ur c'hoari diñs etre 2 c'hoarier pe muioc'h.<br>
Ar c'hoarier kentañ a taol un diñs. Ma vez 1 ne vo gounezet netra gant ar c'hoarier hag e vo tro ar c'hoarier all. Ma vez un niverenn all eget 1 e vez ouzhpennet skor an tro gant an niverenn-se hag ar c'hoarier a c'hell c'hoari en dro pe paouez hag ouzhpennañ skor an tro d'e skor hollek.<br>
Ar c'hoarier kentañ da gaout ur skor uheloc'h pe kevatal da 100 a gounez.<br>
[Liamm wikipedia (saozneg)](https://en.wikipedia.org/wiki/Pig_(dice_game))


```
# Emporzhiet e vez ar modul ankivil
import random
```

```
def taolDins():
    # Ar fonksion taolDins a rank adkas un niver anterin etre 1 ha 6
    return random.randint(1, 6)
```

```
# Gwiriekomp ar fonksion 'taolDins'
for i in range(10):
    print(taolDins())
```

```
def troChoari():
    # Ar fonksion troChoari a ra un tro c'hoari a-bezh evit ur c'hoarier
    # Adkaset e vez skor an tro-mañ (un niver anterin)
    skor_an_tro = 0
    while True:
        respont = input(f"Skor an tro {skor_an_tro}, [R]uilh pe [P]aouez : ")
        choaz = respont[0].lower()
        if choaz == 'r':
            dins = taolDins()
            if dins == 1:
                return 0
            skor_an_tro += dins
        elif choaz == 'p':
            return skor_an_tro
        else:
            print("Ket komprenet... skrivit 'r' evit ruilh pe 'p' evit paouez")
```

```
# Klaskomp implij ar fonksion 'troChoari'
troChoari()
```

```
############## DAROUEKAAT AR VARIENNOÙ ###############
skor_A = 0
skor_B = 0

anv_A = input("Petra eo añv ar c'hoarier kentañ ? ")
anv_B = input("Petra eo añv an eil c'hoarier ? ")


############### AZE E KROG AR C'HOARI ################
while skor_A < 100 and skor_B < 100:
    print(f"Tro da {anv_A}, skor: {skor_A}")
    skor_A += troChoari()
    
    print(f"Tro da {anv_B}, skor: {skor_B}")
    skor_B += troChoari()


################### FIN AR C'HOARI ####################
if skor_A > skor_A:
    print("Gourc'hemennoù " + anv_A)
else:
    print("Gourc'hemennoù " + anv_B)
```
