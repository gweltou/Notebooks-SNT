# Poelladenn Python -- ARNm

Setu ur poelladenn evit ar gelennerien SVD hervez ar poelladennoù Python a c'heller kavout al levr evit ar c'hentañ klas (embanadur Nathan).<br>
Cheñchet em eus un tamm ar c'hod evit ma vefe anatoc'h hag aesoc'h da lenn (ha da gompren). Ouzhpennet em eus fonksionoù evit genel ARNn ankivil, evit mutañ ar sekañsoù bazennoù hag evit keñveriañ daou sekañs.<br><br>

```
# Implijet e vez ur geriadur (framm data 'dict') evit mirout hollad ar c'hodoñsoù
# En ur geriadur e vez liammet pep alc'hwez (ur kodon) gant un talvoud
# (un asid amine)
kodonsou = {
         "UUU": "Phe", "UUC": "Phe", "UUA": "Leu", "UUG": "Leu",
         "UCU": "Ser", "UCC": "Ser", "UCA": "Ser", "UCG": "Ser",
         "UAU": "Tyr", "UAC": "Tyr", "UAA": "STOP", "UAG": "STOP",
         "UGU": "Cys", "UGC": "Cys", "UGA": "STOP", "UGG": "Trp",
         "CUU": "Leu", "CUC": "Leu", "CUA": "Leu", "CUG": "Leu",
         "CCU": "Pro", "CCC": "Pro", "CCA": "Pro", "CCG": "Pro",
         "CAU": "His", "CAC": "His", "CAA": "Gln", "CAG": "Gln",
         "CGU": "Arg", "CGC": "Arg", "CGA": "Arg", "CGG": "Arg",
         "AUU": "Ile", "AUC": "Ile", "AUA": "Ile", "AUG": "Met",
         "ACU": "Thr", "ACC": "Thr", "ACA": "Thr", "ACG": "Thr",
         "AAU": "Asn", "AAC": "Asn", "AAA": "Lys", "AAG": "Lys",
         "AGU": "Ser", "AGC": "Ser", "AGA": "Arg", "AGG": "Arg",
         "GUU": "Val", "GUC": "Val", "GUA": "Val", "GUG": "Val",
         "GCU": "Ala", "GCC": "Ala", "GCA": "Ala", "GCG": "Ala",
         "GAU": "Asp", "GAC": "Asp", "GAA": "Glu", "GAG": "Glu",
         "GGU": "Gly", "GGC": "Gly", "GGA": "Gly", "GGG": "Gly",
        }

```

```
# Ezhomm on o eus ar modul 'random' evit implij ar fonksionoù ankivil
import random
```

```
def arnAnkivil(ment):
    # Ar fonksion 'choices' a adkas un listennad ankivil a ment k (arguzen 'k')
    # diouzh ur poblañs orin (ar chadenn "UCAG" amañ)
    # Ar metod 'join' a stag pep elfenn eus un listenn gant ur chadenn lizherennek
    # (ar chadenn goulo '' amañ)
    # Staget e vez ar sekañs "AUG" e penn kentañ
    return 'AUG' + ''.join(random.choices("UCAG", k=ment))
```

```
# Klaskomp krouiñ un nebeut sekañsoù arn gant ar fonksion arnAnkivil nevez krouet
print(arnAnkivil(100))
print(arnAnkivil(100))
print(arnAnkivil(100))
```

```
def troidigezh(arn):
    # Ar fonksion 'troidigezh' a dro ur sekañs bazennoù (chadenn lizherenek)
    # en ul listennad asid amine diouzh ar geriadur 'kodonsou' bet termenet e
    # penn kentañ ar poelladenn.
    # Adkaset e vez un listenn eus an doare : ['Met', 'Glu', 'Asp, 'Ser', ...]

    # El listenn 'protein' e vo lakaet an asidoù amine bet troet
    protein = []

    plas = 0
    while plas < len(arn)-3:
        # Lennet e vez 3 bazenn eus ar sekañs
        kodon = arn[plas:plas+3]
        # Klasket e vez an asid amine a glot er geriadur 'kodonsou'
        aa = kodonsou[kodon]
        # Paouezet e vez ma vez kavet ur kodon 'STOP'
        if aa == "STOP":
            break
        # Ouzhpennet e vez an asid amine d'ar protein
        # ha aet e vez war-raok eus 3 plas
        protein.append(aa)
        plas += 3
    
    return protein
```

```
# Klaskomp implij ar fonksion 'troidigezh'
arn = arnAnkivil(100)
print(arn)
print(troidigezh(arn))
```

```
def mutaat(arn):
    # Ar fonksion 'mutaat' a gemer un arn (chadenn lizherennek) hag a adkas
    # un arn mutet en un doare ankivil
    mutet = ""
    for bazenn in arn:
        # 10% a probablentez mutatur
        if random.random() < 0.1:
            mutet += random.choice("UCAG")
        else:
            mutet += bazenn
    
    return mutet
```

```
# Klaskomp implij ar fonksion 'mutaat'
arn = arnAnkivil(100)
print(arn)
print(mutaat(arn))
```

```
def kenverian(arn1, arn2):
    # Gant ar fonksion 'kenverian' e vez keñveriet daou sekañs arn evit gwelet
    # en un doare splan an difoc'hoù etrezo
    # Ne vez adkaset talvoud ebet gant ar fonksion-se met skrivet e vez ar daou
    # sekañs war ar skramm (gant ar fonksion 'print') hag ur chadenn lizherennek
    # difoc'h etrezo.
    #
    # Gortozet e vez un diskouel eus ar stumm :
    # AUGCUCGAAACGCUGCGUCAGAGUUAUAA ...
    # ---X---X-----X-X------------- ... (hag all)
    # AUGUUCGCAACGCAGAGUCAGAGUUAUAA ...
    
    dif = ""
    for i in range(min(len(arn1), len(arn2))):
        if arn1[i] == arn2[i]:
            dif += '-'
        else:
            dif += 'X'
    
    print(arn1)
    print(dif)
    print(arn2)
```

```
# Klaskomp implij pep tra
arn = arnAnkivil(100)
arn2 = mutaat(arn)
kenverian(arn, arn2)
print(troidigezh(arn))
print(troidigezh(arn2))
```
