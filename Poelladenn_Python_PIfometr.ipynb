Klasket e vo tostaat eus $\pi$ dre doare Monte Carlo.<br>


Keñveriomp gorreat ur c'helc'h a skin $s$ gant hini ur c'harrez a gostez $2×s$ (diametr ar c'helc'h). Gorreat ar c'harrez a vo brasoc'h evel just peogwir ez eo enklozed ar c'helc'h penn-da-benn er c'harrez. Ped gwech brasoc'h e gwirionez ?<br>
![](http://jakascorner.com/pics/inscribed_circle.png)

Gorread ur c'helc'h a zo $\pi.s^2$ hag hini ar c'harrez enklozus a zo $2 s × 2 s$ (pe $4×s^2$).<br>
Ma vez rannet gorread ar c'helc'h dre hini ar c'harrez e vo kavet neuze :
$${\pi.s^2 \over 4×s^2} = {\pi \over 4}$$
Da lavaret eo ez eo gorread ar c'helc'h $\pi \over 4$ gwech hini ar c'harrez, pe c'hoaz :
$$\pi = 4×{gorread\ ar\ c'helc'h \over gorread\ ar\ c'harrez}$$

M'on eus un doare da muzuliañ gorread ar c'helc'h hag hini ar c'harrez, rannañ an hini gentañ gant an eil ha lieskementiñ ar c'hosiant dre 4 e vefe tu kavout $\pi$.
![](http://jakascorner.com/pics/generate_points_pi.png)

Doare **Monte Carlo** a zo un doare da rakjediñ ar c'hosiant-se en ur implij talvoudoù **dargouezhek**. Bannet e vez ur bern poentoù dargouezhek e-barzh ar c'harrez ha kontet e vez an niver a poentoù a gouezh e-barzh ar c'helch. Ar c'hosiant etre an niver a pentoù er c'helc'h hag an niver hollek a poentoù bet bannet er c'harrez a zo un tostadur eus ar c'hosiant etre ar gorreadoù. Ar muioc'h a poentoù a vo, ar resisoc'h e vo an tostadur.<br><br>
$${gorread\ ar\ c'helc'h \over gorread\ ar\ c'harrez} \simeq {poentoù\ er\ c'helch \over poentoù\ hollek}$$
Neuze
$$\pi \simeq 4×{pentoù\ er\ c'helch \over poentoù\ hollek}$$

Ha penaos gouzout ma gouezh ar poent e-barzh ar c'helc'h ?<br>
N'on eus nemet da jediñ an hed etre ar poent ha kreiz ar c'helc'h. Ma'z eo an hed **nebeutoc'h eget skin** ar c'helc'h e vo ar poent **e-barzh** ar c'helc'h.<br>
Evit jediñ an hed e c'heller implij ar formulenn $hed = \sqrt{x^2 + y^2}$ (ar fonksion grwizienn karrez a vez kavet e Python gant `math.sqrt`) pe, aesoc'h c'hoaz, gant ar fonksion `math.hypot`
