## Ar variennoù
Tu 'zo stokañ **roadennoù** (data) e memor bev an urzhiataer gant variennoù.
Ur **varienn** a zo un anv (dibabet ganeoc'h) liammet gant lodenn eus **memor bev** an urzhiataer ha d'an **talvoud** bet lakaet enno (niverel, lizherennek pe doareoù all).
Anv ur varienn a c'hell bezañ skrivet gant neusforz peseurt lizherenn, sifr pe an arouez _ , da skouer : `x`, `alpha`, `poent_buhez`, `respont1`, `niverLevrBrezhoneg`... met arabat e krogfe anv ur varienn gant ur sifr ! Deoc'h da zibab anvioù anat hag aes da lenn.<br>

```python
oadKatell = 16

# Setu eruet deiz-ha-bloaz Katell, ur bloaz ouzhpenn he deus
oadKatell + 1
print("Oad Katell :", oadKatell)

# Jedet eo bet oad nevez Katell gant jubenner Python met
# n'eo ket bet stoket an disoc'h.
# oadKatell n'eo ket bet kemmet gant ar jedadenn.
# Evit ma vefe miret an oad nevez eo ret kemmañ ar varienn
# a orin, pe krouiñ ur varienn nevez

oadNevez = oadKatell + 1
print("Oad nevez :", oadNevez)
print("Oad Katell :", oadKatell)
# An disoc'h a zo bet lakaet er varienn oadNevez met n'eo
# ket bet chenchet ar varienn a orin c'hoazh

oadKatell = oadKatell + 1
print("Oad Katell :", oadKatell)
```

## Al listennoù

Al listennoù a zo un doare data **kompozet** gant roadennoù all, **finit** hag **urzhiet**.<br>
Ul listenn a c'hell dougen neusforz peseurt niver a roadenn gant doareoù disheñvel.<br>
Termenet e vez un listenn pa vez meur a roadenn da-heul, dispartiet gant **skejoù** ha bevennet gant **krommelloù karrez**.

```python
# Setu ul listenn o tougen roadennoù a doare disheñvel
[5, "glas", -10, print, "WASSUP ?", 3.1415]

# Evit krouiñ ul listenn goulo e vez skrivet krommelloù hep netra en diabarz
listenn_goulo = []
# tu 'zo ivez implij ar fonksion "list"
listenn_goulo_2 = list()

print(listenn_goulo)
print(listenn_goulo_2)
```

Evit ouzhpennañ roadennoù e barz al listenn e vez implijet ar **metod** `append`.<br>
Ur **metod** a zo ur fonksion liammet d'un doare data nemetken. Gant pep doare data e teu alies meur a metodoù. Dizoloet e vint ganeomp tamm-ha-tamm.<br>
Implijet e vez ur metod evel ur fonksion boutin met skrivet e vez e anv da-heul anv ur varienn, gant ur **pik** etrezo.

```python
adn = ['a', 'c', 'c', 'g']
adn = adn + ['c', 'a', 't']
print(adn)
adn = adn + 3 * ['g']
print(adn)
adn *= 2         # Memes tra eget adn = adn * 2

print(adn)
```

### Ar roll `for`

Ar roll `for` (e galleg e vefe laret "*pour chaque*") a servij da peurgas meur a wech ur **blok kod**.<br>
Ur **blok kod** a zo meur linennoù kod da-heul, bet dilec'hiet war an tu kleiz gant un nebeut esaouennoù (4 esaouenn dre boaz). An dilec'hiadenn-se a vez galvet un **endantadur** (*indentation* e galleg hag e saozneg).
Diazezet e vez ar framm roll war un heuliad roadennoù (un listenn pe ur chadenn lizherennek da skouer). E pep **iteradur** (un dro eus ar roll) e vez lakaet an talvoud da-heul en heuliad e barzh ur varienn a vo implijet (pe get) e blok kod ar roll. An niver a **iteradur** eus a roll a vo neuze kevatal da ment an heuliad roadennoù.
Skrivet e vez :
<pre>
for varienn in heuliad:
    blok kod...
</pre>

```python
heuliad = ["unan", "daou", "tri", "daouzek"]
for i in heuliad:
    # E pep tro a roll e vo lakaed un talvoud nevez eus an heuliad
    # e barzh ar varienn i
    print("talvoud 'i' :", i)
```

```python
# Mont a ra en dro gant chadennoù lizherennek ivez
for lizherenn in "abcdefgh":
    print(lizherenn)
```

```python
# Ha setu ur skouer evit sammañ ur listennad rouadennoù asambles
temperadur = [12.5, 16, 10.6, 14.2, 8.8, 23, -1.1, 4.9, 9, 15.7]
hollad = 0
for t in temperadur:
    hollad += t

print("hollad :", hollad)
print("keidenn :", hollad / len(temperadur))
```

E gwirionez e vez kavet ivez ar fonksion `sum()` evit jediñ un hollad.

```python
print("hollad :", sum(temperadur))
print("keidenn :", sum(temperadur) / len(temperadur))
```

### Ar framm diviz `if-elif-else`

Setu ur framm pouezus tre hag a vez kavet e barzh pep yezh programiñ. Gantañ e c'hell cheñch red ur program hervez un **amplegad** (pe kondision) bennak. Heptañ ur program ne vefe nemet un heuliad a stadamantoù finit ha statik.<br>
Gant ar framm diviz, ma vez gwiriekaet un amplegad e vo peurgaset un blok kod bennak ha mod all (ma vez faos disoc'h an amplegad) e vo peurgaset ur blok kod all.<br>
Setu penaos e vez skrivet :
<pre>
if amplegad:
    blok kod peurgaset ma vez gwir an amplegad
else:
    blok kod peurgaset ma vez faos an amplegad
</pre>

Ar stadamant `else` a zo diret. Tu vefe ober gant ar stadamant `if` nemetken.<br>
M'o peus c'hoant gwiriekaat meur a amplegadoù ez eus tu ouzhpennañ stadamantoù `elif` (berradenn "else if").<br>
<pre>
if amplegad1:
    blok kod peurgaset ma vez gwir an amplegad 1
elif amplegad2:
    blok kod peurgaset ma vez gwir an amplegad 2
elif amplegad3:
    blok kod peurgaset ma vez gwir an amplegad 3
else:
    blok kod peurgaset ma vez faos tout an amplegadoù

Un **amplegad** a zo ur stadamant gant un disoc'h gwir pe gaou nemetken (`True` pe `False` e Python).<br>
Ur **c'henveriadenn** etre div talvoud a zo un doare amplegadenn.

```python
notennoù_jobig = [14.5, 13, 6, 8.5, 14, 9.5, 10, 12.5, 16]
notennoù_mat = []
notennoù_fall = []
bevenn = 10         # Cheñchit ar bevenn evit gwellet an disoc'h

notennoù_jobig.sort()    # Kenkoulz lakaat an notennoù en urzh da gentañ

for notenn in notennoù_jobig:
    if notenn >= bevenn:
        notennoù_mat.append(notenn)
    else:
        notennoù_fall.append(notenn)

print("Notennoù mat :", notennoù_mat)
print("Notennoù fall :", notennoù_fall)
```

### Ar roll `while`

Ar roll `while` a vez implijet evit peurgas un blok kod adare, keit a ma vez asantet d'un **amplegad**. Ma vez gwir an amplegad e vo kendalc'het ar roll. Paouez a raio ar roll nemet ma vez faos an amplegad. Na zisoñjit ket endantañ ar blok kod !<br>
Skrivet e vez :
<pre>
while amplegad:
    blok kod...
</pre>

```python
a = 1
while a <= 1000:    # Keit a ma'z eo 'a' bihanoc'h pe kevatal da 1000...
    a = a*2

print(a)
```

Ar fonksion `input` a enkas ur talvoud goulennet d'an implijer. An etrefas simplañ eo etre ar program hag an implijer.<br>
Diwallit gant ar pezh a vez adkaset gant ar fonksion `input`. Atav e vo ur **chadenn lizherennek**. M'o peus ezhomm goulenn un niverenn e vo ret deoc'h amdreiñ ar roadenn adkaset gant ar fonksion `int` pe `float`.

```python
# Gwiriekaat ar pez a vez respontet gant an implijer

r = input("Soubenn pe fritez ? ")
while r not in ["soubenn", "fritez"]:
    print("N'eo ket", r, "er roll boued...")
    r = input("Soubenn pe fritez ? ")

print(r, "a vo debret fenoz !")
```

```python
# Ar roll mañ ne paouezo morse. Paouezit anezhi "dre-dorn" gant ar bouton "stop"
# "pass" a zo ur stadamant ispisial avit lavar da Python ober netra
while True:
    pass
```

```python
# To zo mont er maez eus ur framm roll neusforz peur gant ar stadamant 'break'
i=0
while True:
    if i==100:
        print("Paouezet e vez ar roll")
        break
    i += 1

print("Talvoud 'i' :", i)
```
